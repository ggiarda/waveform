# Data Analysis

```python
import pysr
from pysr import PySRRegressor

import csv
import numpy as np
import matplotlib.pyplot as plt

from sympy import Piecewise
from IPython.display import display, Math
```

## Data Reading

```python
# Specify the CSV file path
csv_file_path = 'time_data.csv'

# Read from CSV and create NumPy arrays
data = np.genfromtxt(csv_file_path, delimiter=',', skip_header=1, dtype=None)

# Extract columns

X = data[:, 0]
y = data[:, 1]
q = data[:, 2]

Xdraw = X

# Use this if you don't want to use q as a feature
        
X = X.reshape(-1, 1)

# Use this if you want to use q as a feature

#X = np.column_stack((X, q))
```

## Model Loading

```python
model_time = PySRRegressor.from_file("hall_of_fame_2023-12-27_104050.869.pkl")
```

## Analysis

```python
ypred = model_time.predict(X)
```

```python
plt.plot(Xdraw, y, c = "black", label = "Data")
plt.plot(Xdraw, ypred, c = "red", label = "Prediction")
plt.legend()
plt.show()
```
